```python
#####讀取airline資料
airline = pd.read_csv('D:\\dataset\\data\\airline.csv',encoding = 'utf-8')
airline.info()

airline.sample(5)

airline.describe()


###變數轉換
#1.group_id
#2.go back : 兩種可能性，轉成0或1
def recode_go_back(value):
    if value == "去程":
        return 0
    elif value == "回程":
        return 1
    
airline['go_back_recode'] = airline['go_back'].apply(recode_go_back).astype('category')


#3.fly time : 轉成日期格式

def get_year(value):
    return value.year

def get_month(value):
    return value.month
    
def get_day(value):
    return value.day

airline['fly_time'] = pd.to_datetime(airline['fly_time'])
airline['fly_month'] = airline['fly_time'].apply(get_month).astype('category')
airline['fly_year'] = airline['fly_time'].apply(get_year).astype('category')
airline['fly_day'] = airline['fly_time'].apply(get_day).astype('category')


#4.src_airport : 轉成204個類別
def get_airport(value):
    return value[0:3]

airline['src_airport_recode'] = airline['src_airport'].apply(get_airport).astype('category')


#5.arrive time :轉成日期格式
airline['arrive_time'] = pd.to_datetime(airline['arrive_time'])
airline['arrive_month'] = airline['arrive_time'].apply(get_month).astype('category')
airline['arrive_year'] = airline['arrive_time'].apply(get_year).astype('category')
airline['arrive_day'] = airline['arrive_time'].apply(get_day).astype('category')

airline.info()


#6.dst airport : 
airline['dst_airport_recode'] = airline['dst_airport'].apply(get_airport).astype('category')



```

```python
#begin_date/order_date:日期變數
#group_id,sub_line,area,source1,source2,unit:類別變數
#須改為類別變數:order_id,deal or not
#days,price,people_amount:數值變數


#turn begin_date and order_date to date format

feature_data['begin_date'] = pd.to_datetime(feature_data['begin_date'])
feature_data['order_date'] = pd.to_datetime(feature_data['order_date'])

###看變數的描述性統計
#print('category vars')
#print(feature_data[['group_id','sub_line','area','begin_date','order_date','order_id','source_1','source_2','unit','deal_or_not']].describe())

#print('')
#print('continuous vars')
#print(feature_data.describe())



###類別變數的recoding

##source_1的recoding
#原本變數的類別，每個類別各有幾次
#feature_df['source_1'].value_counts()


def recode_source_1(value):
    if value == "src1_value_1":
        return 1
    elif value == "src1_value_2":
        return 2
    elif value == "src1_value_3" :
        return 3

feature_data['source_1_recode'] = feature_data['source_1'].apply(recode_source_1)

#看recode後的source_1是不是OK
#feature_data['source_1_recode'].value_counts(sort=False)

##source_2的recoding

#feature_data['source_2'].value_counts()

#
def recode_source_2(value):
    if value == "src2_value_1":
        return 1
    elif value == "src2_value_2":
        return 2
    elif value == "src2_value_3":
        return 3
    elif value == "src2_value_4":
        return 4

feature_data['source_2_recode'] = feature_data['source_2'].apply(recode_source_2)
#feature_data['source_2_recode'].value_counts(sort = False)



##sub_line的recoding:抓取_後面的數字當recode的值
def recode_subline(value):
    data = value.split('_')[2]
    return data
    
    
    

feature_data['subline_recode'] = feature_data['sub_line'].apply(recode_subline)
#print(feature_data['subline_recode'].value_counts(sort= True))
#print(feature_data['sub_line'].value_counts(sort= True))




##area的recoding
def recode_area(value):
    data = value.split('_')[2]
    return data
    
feature_data['area_recode'] = feature_data['area'].apply(recode_area)


##unit的recoding
def recode_unit(value):
    data = value.split('_')[2]
    return data

feature_data['unit_recode'] = feature_data['unit'].apply(recode_unit)

##order_date的recoding

def get_year(value):
    return value.year

def get_month(value):
    return value.month
    
def get_day(value):
    return value.day
    
feature_data['order_month'] = feature_data['order_date'].apply(get_month)
feature_data['order_year']  = feature_data['order_date'].apply(get_year)
feature_data['order_day'] = feature_data['order_date'].apply(get_day)

##begin_date的recoding
    
feature_data['begin_month'] = feature_data['begin_date'].apply(get_month)
feature_data['begin_year']  = feature_data['begin_date'].apply(get_year)
feature_data['begin_day'] = feature_data['begin_date'].apply(get_day)






#更改變數類別
feature_data['order_id'] = feature_data['order_id'].astype('category') 
feature_data['subline_recode'] = feature_data['subline_recode'].astype('category')
feature_data['source_1_recode'] = feature_data['source_1_recode'].astype('category')
feature_data['source_2_recode'] = feature_data['source_2_recode'].astype('category')
feature_data['area_recode'] = feature_data['area_recode'].astype('category')
feature_data['unit_recode'] = feature_data['unit_recode'].astype('category')
feature_data['order_month'] = feature_data['order_month'].astype('category')
feature_data['order_year'] = feature_data['order_year'].astype('category')
feature_data['order_day'] = feature_data['order_day'].astype('category')
feature_data['begin_month'] = feature_data['begin_month'].astype('category')
feature_data['begin_year'] = feature_data['begin_year'].astype('category')
feature_data['begin_day'] = feature_data['begin_day'].astype('category')


feature_data.info()
```

```python
#####建立隨機森林的預測模型

from sklearn import cross_validation, ensemble, preprocessing, metrics

# 創造 dummy variables
#label_encoder = preprocessing.LabelEncoder()
#encoded_Sex = label_encoder.fit_transform(titanic_train["Sex"])

# 建立訓練與測試資料
train_X = pd.DataFrame([modeling_data["people_amount"],
                        modeling_data["days"],
                        modeling_data["price"],
                        modeling_data["subline_recode"],
                        modeling_data["source_1_recode"],
                        modeling_data["source_2_recode"],
                        modeling_data["area_recode"],
                        modeling_data["unit_recode"],
                        modeling_data["order_month"],
                        modeling_data["order_year"],
                        modeling_data["order_day"],
                        modeling_data["begin_month"],
                        modeling_data["begin_year"],
                        modeling_data["begin_day"]]).T

train_y = modeling_data["deal_or_not"]


train_X, test_X, train_y, test_y = cross_validation.train_test_split(train_X, train_y, test_size = 0.2)
print(train_X)
print(train_y)

# 建立 random forest 模型
forest = ensemble.RandomForestClassifier(n_estimators = 100)
forest_fit = forest.fit(train_X, train_y)



# 預測
test_y_predicted = forest.predict(test_X)

# y是"數值型"績效
#accuracy = metrics.accuracy_score(test_y, test_y_predicted)
#print(accuracy)

# y是"類別型"績效
fpr, tpr, thresholds = metrics.roc_curve(test_y, test_y_predicted)
auc = metrics.auc(fpr, tpr)
print(auc)
```
